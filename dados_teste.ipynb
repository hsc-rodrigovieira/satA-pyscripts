```python
import pandas as pd
import random

def gerar_dados_hospitalares():
    # Configuração básica
    anos = [2023,2024]
    # meses = ["Jan", "Fev", "Mar", "Abr", "Maio", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]
    meses = list(range(1,13))
    #hospitais = [{"$oid":"6745d7873f4e39e161319575"}, {"$oid":"6745d7d93f4e39e161319576"}]
    hospitais = [9876543, 1234567]
    faixas_etarias = ["neo_precoce", "neo_tardio", "pedi", "ad", "idoso"]
    faixas_etarias2 = ["neo", "pedi", "ad"]
    tipos_internacao = ["cli", "cir"]
    tipos_unidades = ["ui", "uti"]
    tipos_pacientes = ["cli", "cir_orto", "cir_nao_orto"]
    dados = []

    for hospital in hospitais:
        for ano in anos:
            for mes in meses:
                # Indicador 1: Taxa de partos
                partos_vaginais = random.randint(50, 150)
                partos_cesareos = random.randint(50, 100)

                # Indicador 2: Reinternações
                saidas_clinicas_anterior = random.randint(300, 600)
                saidas_cirurgicas_anterior = random.randint(200, 400)
                reinternacoes_clinicas = random.randint(10, 50)
                reinternacoes_cirurgicas = random.randint(5, 30)            

                # Indicador 3: Taxa PCR
                pacientes_dia = random.randint(4000, 6000)
                pcr_eventos = random.randint(0, 5)

                # Indicador 4: Taxa de Mortalidade
                # Indicador 5: Tempo médio de internação
                dados_estratificados = {}
                for tipo in tipos_internacao:
                    for faixa_etaria in faixas_etarias:                
                        saidas = random.randint(100, 300)
                        obitos = random.randint(1, 30)
                        pacientes_dia = random.randint(500, 1000)

                        dados_estratificados[f"{tipo}_{faixa_etaria}_saidas"] = saidas
                        dados_estratificados[f"{tipo}_{faixa_etaria}_obitos"] = obitos    
                        dados_estratificados[f"{tipo}_{faixa_etaria}_pacientes_dia"] = pacientes_dia

                # Indicador 6: Tempo médio de permanência na emergência
                total_pacientes_emergencia = random.randint(100, 300)
                total_tempo_permanencia_emergencia_hr = random.randint(300, 900)  # Em horas 

                # Indicador 7: Tempo médio de espera na emergência para 1º atendimento
                pacientes_emergencia_nivel2 = random.randint(20, 100)
                pacientes_emergencia_nivel3 = random.randint(50, 150)
                tempo_total_emergencia_nivel2_min = random.randint(200, 600)  # Em minutos
                tempo_total_emergencia_nivel3_min = random.randint(1500, 3000)  # Em minutos  

                # Indicador 8: Taxa de início de antibiótico intravenoso profilático
                total_cirurgias_limpas = random.randint(50, 150)
                cirurgias_com_antibiotico = random.randint(40, total_cirurgias_limpas)   

                # Indicador 9: Taxa de infecção de sítio cirúrgico em cirurgia limpa
                total_cirurgias_limpas_mes_anterior = random.randint(50, 150)
                total_infeccoes = random.randint(0, 5)

                # Indicador 10: Densidade de incidência de infecção de corrente sanguínea
                dados_infeccao = {}
                for faixa in faixas_etarias2:
                    for tipo in tipos_unidades:
                        cvc_dia = random.randint(200, 500)
                        infeccoes_cvc = random.randint(0, 10)                    
                        dados_infeccao[f"{tipo}_{faixa}_cvc_dia"] = cvc_dia
                        dados_infeccao[f"{tipo}_{faixa}_infec"] = infeccoes_cvc      

                # Indicador 11: ITU associada a CVD
                dados_itu = {}
                for faixa in faixas_etarias2:
                    for unidade in tipos_unidades:
                        cvd_dia = random.randint(50, 300)
                        itu_cvd = random.randint(0, 10)                    
                        dados_itu[f"{unidade}_{faixa}_cvd_dia"] = cvd_dia
                        dados_itu[f"{unidade}_{faixa}_itu"] = itu_cvd                    

                # Indicador 12: Profilaxia de TEV
                dados_tev = {}
                for tipo in tipos_pacientes:
                    total_pacientes = random.randint(50, 200)
                    pacientes_profilaxia = random.randint(40, total_pacientes)
                    dados_tev[f"{tipo}_total_pacientes"] = total_pacientes
                    dados_tev[f"{tipo}_profilaxia"] = pacientes_profilaxia

                # Indicador 13: Quedas com lesão
                # pacientes_dia = random.randint(4000, 6000)
                quedas_com_dano = random.randint(0, 10)

                # Indicador 14: Eventos sentinela
                # pacientes_dia = random.randint(4000, 6000)
                eventos_sentinela = random.randint(0, 5)

                # Consolidar os dados
                dados.append({
                    "organization_cnes": hospital,
                    "year": ano,
                    "month": mes,
                    "partos_vaginais": partos_vaginais,
                    "partos_cesareos": partos_cesareos,
                    "saidas_clinicas_anterior": saidas_clinicas_anterior,
                    "saidas_cirurgicas_anterior": saidas_cirurgicas_anterior,
                    "reinternacoes_clinicas": reinternacoes_clinicas,
                    "reinternacoes_cirurgicas": reinternacoes_cirurgicas,  
                    "pacientes_dia": pacientes_dia,
                    "pcr_eventos": pcr_eventos,
                    **dados_estratificados,
                    "total_pacientes_emergencia": total_pacientes_emergencia,
                    "total_tempo_permanencia_emergencia_hr": total_tempo_permanencia_emergencia_hr,
                    "pacientes_emergencia_nivel2": pacientes_emergencia_nivel2,
                    "pacientes_emergencia_nivel3": pacientes_emergencia_nivel3,
                    "tempo_total_emergencia_nivel2_min": tempo_total_emergencia_nivel2_min,
                    "tempo_total_emergencia_nivel3_min": tempo_total_emergencia_nivel3_min,
                    "total_cirurgias_limpas": total_cirurgias_limpas,
                    "cirurgias_com_antibiotico": cirurgias_com_antibiotico,
                    "total_cirurgias_limpas_anterior": total_cirurgias_limpas_mes_anterior,
                    "total_infeccoes": total_infeccoes,
                    **dados_infeccao,
                    **dados_itu,
                    **dados_tev,
                    "quedas_com_dano": quedas_com_dano,
                    "eventos_sentinela": eventos_sentinela
                })

    return pd.DataFrame(dados)

# Gerar os dados e salvar em CSV
dados_completos = gerar_dados_hospitalares()
caminho_arquivo = "dados_hospitalares_completos.csv"
dados_completos.to_csv(caminho_arquivo, index=False)

print(f"Dados gerados e salvos em: {caminho_arquivo}")
dados_completos.head()
```
